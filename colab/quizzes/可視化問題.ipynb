## 1問目

```python
f = pd.read_csv('C:/Users/isami/Basic-Python/colab/quizzes/matplotlib_quiz/hist_data.csv',names = ['aデータ','bデータ'])
adata = f['aデータ'][1:].astype('float64')
bdata = f['bデータ'][1:].astype('float64')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
plt.xticks([0,10,20,30,40,50])
plt.yticks([0,5,10,15,20,25])
plt.hist(adata, alpha = 0.4, bins = 30, color = 'goldenrod')
plt.hist(bdata, alpha = 0.4, bins = 30, color = 'darkturquoise')
plt.legend(['aデータ','bデータ'], loc = 'upper right')
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
g = pd.read_csv('C:/Users/isami/Basic-Python/colab/quizzes/matplotlib_quiz/scatter_data.csv', names = ['length','width','class'])
scatter_list = g[1:].astype('float64')
class0 = scatter_list[0:50]
class1 = scatter_list[50:100]
class2 = scatter_list[100:150]

plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.xlabel('length(cm)')
plt.ylabel('width(cm)')
plt.grid(True)
plt.scatter(class0['length'],class0['width'], c  ='red')
plt.scatter(class1['length'],class1['width'], c = 'blue')
plt.scatter(class2['length'],class2['width'], c = 'green')
plt.legend(['class0','class1','class2'], loc = 'upper right')
```

```python
h = pd.read_csv('C:/Users/isami/Basic-Python/colab/quizzes/matplotlib_quiz/line_data.csv', names = ['期間','件数','費用'])
data_list = h[1:]

figure  = plt.figure(figsize = (6,10))
plt.subplots_adjust(hspace = 0.5)
axes = figure.add_subplot(2,1,1)
x = [1,2,3,4,5,6,7,8]
axes.set_xticks(x,data_list['期間'], rotation = 30)
axes.set_title('期間ごとの件数の推移')
axes.set_xlabel('期間')
axes.set_ylabel('件数')
height = np.array(data_list['件数'].astype('int64'))
axes.plot(x,height,marker = 'o')
axes.grid(True)

axes2 = figure.add_subplot(2,1,2)
axes2.set_xticks(x,data_list['期間'], rotation = 30)
axes2.set_title('期間ごとの費用の推移')
axes2.set_xlabel('期間')
axes2.set_ylabel('費用(円)')
axes2.bar(x,data_list['費用'].astype('int64'))
axes2.grid(True)
```
